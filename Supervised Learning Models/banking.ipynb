```python
df_encoded = pd.get_dummies(df, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome' ])
```

```python
df_encoded['y'] = df_encoded['y'].map({'yes': 1, 'no': 0})
```

```python
df_encoded.dtypes
```

```python
display(df_encoded.corr())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = df_encoded.drop('y',axis=1)
y = df_encoded['y']
```

```python
X_test,X_train,y_test,y_train = train_test_split(X , y ,test_size = 0.33 ,random_state=42)
```

```python
numerical_col = X_train.select_dtypes(include=['float64','int64']).columns

sc=StandardScaler()
X_train[numerical_col]= sc.fit_transform(X_train[numerical_col])
X_test[numerical_col]=sc.transform(X_test[numerical_col])
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr_pred=lr.predict(X_test)
print(classification_report(y_test,lr_pred))
```

```python
y_pred=lr.predict(X_test)
acc = accuracy_score(y_test,y_pred)
print("Accuracy Score", round(acc * 100 , 2), "%")
```
