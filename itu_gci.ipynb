```python
# One-hot encode categorical columns
categorical_cols = ['REF_AREA', 'REF_AREA_LABEL', 'INDICATOR', 'INDICATOR_LABEL', 'UNIT_MEASURE_LABEL', 'TIME_PERIOD']
X = pd.get_dummies(df[categorical_cols], drop_first=True)
y = df['OBS_VALUE']
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
numerical_cols = X_train.select_dtypes(include=['float64', 'int64']).columns


sc=StandardScaler()
X_train[numerical_cols] = sc.fit_transform(X_train[numerical_cols])

X_test[numerical_cols]= sc.transform(X_test[numerical_cols])

print(X_train[numerical_cols].dtypes)
print(X_train[numerical_cols].head())



```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)

print("Linear Regression:")
print(f"  - RMSE: {mean_squared_error(y_test, lr_preds, squared=False):.2f}")
print(f"  - R²: {r2_score(y_test, lr_preds):.3f}\n")
```

```python
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)
dt_preds = dt.predict(X_test)

print("Decision Tree:")
print(f"  - RMSE: {mean_squared_error(y_test, dt_preds, squared=False):.2f}")
print(f"  - R²: {r2_score(y_test, dt_preds):.3f}\n")
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)

print(" Random Forest:")
print(f"  - RMSE: {mean_squared_error(y_test, rf_preds, squared=False):.2f}")
print(f"  - R²: {r2_score(y_test, rf_preds):.3f}\n")
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
r2 = r2_score(y_test, y_pred_rf)
print(f"R² Score (Model Accuracy Equivalent): {r2:.3f}")
```
